```python
results = requests.get(url).json()
results
```

```python
#Asigna la parte relevante de JSON a lugares
venues=results["response"]["venues"]

#Transforma los lugares en un marco de datos

dataframe=pd.json_normalize(venues)
dataframe.head()
```

```python
#Mantén solo las columnas que incluyan el nombre del lugar, y cualquier cosas que esté asociada con la localización
filtered_columns =["name","categories"]+[col for col in dataframe.columns if col.startswith("location.")]+["id"]
dataframe_filtered = dataframe.loc[:, filtered_columns]

#Función que extrae la categoría del lugar

def get_category_type(row):
    try:
        categories_list = row["categories"]
    except:
        categories_list= row["venue.categories"]
    if len (categories_list)==0:
       return None
    else:
       return categories_list[0]["name"]

#Filtra la categoría por cada fila
dataframe_filtered["categories"] = dataframe_filtered.apply(get_category_type, axis=1)
```

```python
#Limpía los nombres de la columna manteniendo solo el último término
dataframe_filtered.columns =[column.split(".")[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
venues_map = folium.Map(location=[latitude,longitude], zoom_start=13) #genera una mapa centrado
#añade una marca de círculo azul claro para representar el Hotel Conrad
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color="green",
    popup="Conrad Hotel",
    fill=True,
    fill_color="green",
    fill_opacity=0.6
).add_to(venues_map)

#Añade los restaurantes italianos como marcas de círculos de color rosa salmón:

for lat, lng, label in zip(dataframe_filtered.lat, dataframe_filtered.lng, dataframe_filtered.categories):
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        color="black",
        popup=label,
        fill=True,
        fill_color="lightsalmon",
        fill_opacity=0.6
).add_to(venues_map)

#Mostrar el mapa

venues_map
```
